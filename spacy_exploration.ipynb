```python
import spacy
from spacy.matcher import PhraseMatcher
from spacy.matcher import Matcher
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
text = open('dataset/plain_text/text_1.txt', 'r').read()
```

```python
doc = nlp(text)
```

```python
sentence_spans = list(doc.sents)
spacy.displacy.serve(sentence_spans, style="dep")
```

```python
token_example = doc[0]
print(token_example.ent_type_)
print(token_example.pos_)
print(token_example.tag_)
print(token_example.vector)
```

```python
phrase_matcher = PhraseMatcher(nlp.vocab)
phrase_matcher.add("TEST", None, nlp('nairobi'), nlp('NAIROBI'))
matches = phrase_matcher(doc)
matches
```

```python
matches = Matcher(nlp.vocab)
text = 'National Convention Assembly'
doc = nlp(text)
matches.add("ORG", None, [{'POS':'PROPN', 'OP':'+'}, {'TEXT':'Assembly'}])
matches = matches(doc)
matches
```

```python
from ipymarkup import show_box_markup
```

```python
def i_to_idx(matches, doc):
    spans = set()
    for match in matches:
        start_i = match[1]
        end_i = match[2]
        span = doc[start_i:end_i]
        spans.add((span[0].idx, span[0].idx + len(span.text)))
    return spans
```

```python
show_box_markup(doc.text, i_to_idx(matches, doc))
```
